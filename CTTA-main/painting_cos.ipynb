```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
"""target domain数据集在每个BN层的mean"""
with open("D:\Desktop\painting\cifar100\cifar100_gaussian_all_target_mean_per_layer.pkl","rb") as  f:
    data1 = torch.load(f,map_location=torch.device("cpu"))

"""target domain数据集在每个BN层的var"""
with open("D:\Desktop\painting\cifar100\cifar100_gaussian_all_target_var_per_layer.pkl","rb") as  f:
    data2 = torch.load(f,map_location=torch.device("cpu"))
```

```python
"""target domain当前batch1数据集在每个BN层的mean（通过第一层的KL修改所有层的mean）"""
with open("D:\Desktop\painting\cifar100\gaussian_noise_batch3_mean_per_layer_by_cov1.pkl","rb") as  f:
    data3 = torch.load(f,map_location=torch.device("cpu"))

"""target domain当前batch1数据集在每个BN层的var（通过第一层的KL修改所有层的var）"""
with open("D:\Desktop\painting\cifar100\gaussian_noise_batch3_var_per_layer_by_cov1.pkl","rb") as  f:
    data4 = torch.load(f,map_location=torch.device("cpu"))
```

```python
"""target domain当前batch1数据集在每个BN层的mean（计算每一层的KL分别修改所有层的mean）"""
with open("D:\Desktop\painting\cifar100\gaussian_noise_batch3_mean_per_layer_by_cov1_S.pkl","rb") as  f:
    data5 = torch.load(f,map_location=torch.device("cpu"))

"""target domain当前batch1数据集在每个BN层的var（计算每一层的KL分别修改所有层的var）"""
with open("D:\Desktop\painting\cifar100\gaussian_noise_batch3_var_per_layer_by_cov1_S.pkl","rb") as  f:
    data6 = torch.load(f,map_location=torch.device("cpu"))
```

```python
dis1_1 = []
dis1_2 = []
dis1_3 = []
for idx in range(int(len(data3)/3)):
    sim1 = F.cosine_similarity(data1[idx], data3[idx], dim=0)
    sim2 = F.cosine_similarity(data1[idx], data3[idx+31], dim=0)
    sim3 = F.cosine_similarity(data1[idx], data3[idx+31+31], dim=0)
    dis1_1.append(sim1.item())
    dis1_2.append(sim2.item())
    dis1_3.append(sim3.item())
print(dis1_1,dis1_2,dis1_3)
```

```python
import pandas as pd

"""mean"""
def plot_line_chart(dis1_1,dis1_2,dis1_3,dis3_1,dis3_2,dis3_3):
    x1 = list(range(len(dis1_1)))  # 创建x轴数据
    # y1 = 1/3 * (dis1_1 + dis1_2 + dis1_3)                 # 创建y轴数据，假设data是你的列表数据
    x3 = list(range(len(dis3_1)))  # 创建x轴数据
    # y3 = 1/3 * (dis3_1 + dis3_2 + dis3_3)                   # 创建y轴数据，假设data是你的列表数据
    y1 = []
    y3 = []
    for idx in range(len(dis1_1)):
        y1.append((dis1_1[idx] + dis1_2[idx] + dis1_3[idx])/3.0)
        y3.append((dis3_1[idx] + dis3_2[idx] + dis3_3[idx])/3.0)
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams["axes.unicode_minus"] = False
    plt.scatter(x1, y1, label="通过计算cov1的KL修改每层的动量", color="red")             # 绘制折线图
    plt.scatter(x3, y3, label="分别计算每层的KL修改每层的动量", color="blue")             # 绘制折线图
    plt.xlabel('第 th 个BN层')          # 设置x轴标签
    plt.ylabel('cosine值')          # 设置y轴标签
    plt.title('均值')        # 设置图表标题
    plt.legend()
    plt.show()                 # 显示图表
    df1 = pd.DataFrame(y1)
    df1.to_excel("D:\Desktop\y1.xlsx", index=False)
    df3 = pd.DataFrame(y3)
    df3.to_excel("D:\Desktop\y3.xlsx", index=False)
# 调用函数绘制折线图
plot_line_chart(dis1_1,dis1_2,dis1_3,dis3_1,dis3_2,dis3_3)
```

```python
"""mean+var"""
def plot_line_chart(dis1_1,dis1_2,dis1_3,dis3_1,dis3_2,dis3_3,dis2_1,dis2_2,dis2_3,dis4_1,dis4_2,dis4_3):
    x = list(range(len(dis1_1)))  # 创建x轴数据
    x_ = list(range(len(dis3_1)))  # 创建x轴数据
    y = []
    y_ = []
    for idx in range(len(dis1_1)):
        y.append((dis1_1[idx] + dis1_2[idx] + dis1_3[idx] + dis2_1[idx] + dis2_2[idx] + dis2_3[idx])/6.0)                  # 创建y轴数据，假设data是你的列表数据
        y_.append((dis3_1[idx] + dis3_2[idx] + dis3_3[idx] + dis4_1[idx] + dis4_2[idx] + dis4_3[idx])/6.0)                # 创建y轴数据，假设data是你的列表数据
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams["axes.unicode_minus"] = False
    plt.scatter(x, y, label="通过计算cov1的KL修改每层的动量", color="red")             # 绘制折线图
    plt.scatter(x_, y_, label="分别计算每层的KL修改每层的动量", color="blue")             # 绘制折线图
    plt.xlabel('第 th 个BN层')          # 设置x轴标签
    plt.ylabel('cosine值')          # 设置y轴标签
    plt.title('均值+方差')        # 设置图表标题
    plt.legend()
    plt.show()                 # 显示图表
    df5 = pd.DataFrame(y)
    df5.to_excel("D:\Desktop\y5.xlsx", index=False)
    df6 = pd.DataFrame(y_)
    df6.to_excel("D:\Desktop\y6.xlsx", index=False)
# 调用函数绘制折线图
plot_line_chart(dis1_1,dis1_2,dis1_3,dis3_1,dis3_2,dis3_3,dis2_1,dis2_2,dis2_3,dis4_1,dis4_2,dis4_3)
```
